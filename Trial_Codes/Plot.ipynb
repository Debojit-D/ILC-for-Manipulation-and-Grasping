```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Path to the CSV file
file_path = "/home/debojit/data_01/trajectory_data/te_data.csv"

# Read the CSV file into a DataFrame
df = pd.read_csv(file_path, skiprows=1, header=None, names=['TE_x', 'TE_y'])

# Convert string representations of lists to actual lists
df['TE_x'] = df['TE_x'].apply(lambda x: np.array(eval(x)))
df['TE_y'] = df['TE_y'].apply(lambda x: np.array(eval(x)))

# Calculate L2 norm of task space error for each trial
df['L2_norm'] = np.sqrt((df['TE_x'].apply(np.linalg.norm))**2 + (df['TE_y'].apply(np.linalg.norm))**2)

# Plot L2 norm of task space error against trial number
plt.plot(df.index, df['L2_norm'])
plt.xlabel('Trial Number')
plt.ylabel('L2 Norm of Task Space Error')
plt.title('L2 Norm of Task Space Error vs Trial Number')
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Path to the CSV file
file_path = "/home/debojit/data_01/trajectory_data/all_trials_data.csv"

# Read the CSV file into a DataFrame
df = pd.read_csv(file_path)

# Initialize an empty list to store modified time values
modified_time = []

# Iterate through each trial
for trial in df['Trial'].unique():
    # Get the data for the current trial
    trial_data = df[df['Trial'] == trial]
    # Get the minimum time value for the current trial
    min_time = trial_data['Time'].min()
    # Subtract the minimum time value from all time values in the current trial
    modified_time.extend(trial_data['Time'] - min_time)

# Add the modified time values as a new column to the DataFrame
df['Modified_Time'] = modified_time

# Filter the DataFrame for trials 21 to 27
trials_data = df[df['Trial'].isin([f'TN{i}' for i in range(20, 26)])]

# Create a 2x3 grid of subplots
fig, axs = plt.subplots(2, 3, figsize=(15, 10))

# Iterate through each trial and plot 'Fx' and 'Fy' forces against modified time
for i, trial in enumerate(trials_data['Trial'].unique()):
    trial_data = trials_data[trials_data['Trial'] == trial]
    row = i // 3
    col = i % 3
    axs[row, col].plot(trial_data['Modified_Time'], trial_data['Fx'], label=f'{trial} Fx')
    axs[row, col].plot(trial_data['Modified_Time'], trial_data['Fy'], label=f'{trial} Fy')
    axs[row, col].set_xlabel('Time (Relative to Trial Start)')
    axs[row, col].set_ylabel('Force')
    axs[row, col].set_title(f'Fx and Fy Forces vs Time ({trial})')
    axs[row, col].legend()
    axs[row, col].grid(True)

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Path to the CSV file
file_path = "/home/debojit/data_01/trajectory_data/all_trials_data.csv"

# Read the CSV file into a DataFrame
df = pd.read_csv(file_path)

# Initialize an empty list to store modified time values
modified_time = []

# Iterate through each trial
for trial in df['Trial'].unique():
    # Get the data for the current trial
    trial_data = df[df['Trial'] == trial]
    # Get the minimum time value for the current trial
    min_time = trial_data['Time'].min()
    # Subtract the minimum time value from all time values in the current trial
    modified_time.extend(trial_data['Time'] - min_time)

# Add the modified time values as a new column to the DataFrame
df['Modified_Time'] = modified_time

# Filter the DataFrame for trials 21 to 27
trials_data = df[df['Trial'].isin([f'TN{i}' for i in range(20, 26)])]

# Create a 2x3 grid of subplots
fig, axs = plt.subplots(2, 3, figsize=(15, 10))

# Iterate through each trial and plot 'Fx' and 'Fy' forces against modified time
for i, trial in enumerate(trials_data['Trial'].unique()):
    trial_data = trials_data[trials_data['Trial'] == trial]
    row = i // 3
    col = i % 3
    axs[row, col].plot(trial_data['Modified_Time'], trial_data['Fx'], label=f'{trial} Fx', alpha=0.5)
    axs[row, col].plot(trial_data['Modified_Time'], trial_data['Fy'], label=f'{trial} Fy', alpha=0.5)
    
    # Smoothed plots using a moving average (window size = 5)
    axs[row, col].plot(trial_data['Modified_Time'], trial_data['Fx'].rolling(window=7).mean(), label=f'{trial} Fx Smoothed', color='blue')
    axs[row, col].plot(trial_data['Modified_Time'], trial_data['Fy'].rolling(window=7).mean(), label=f'{trial} Fy Smoothed', color='orange')
    
    axs[row, col].set_xlabel('Time (Relative to Trial Start)')
    axs[row, col].set_ylabel('Force')
    axs[row, col].set_title(f'Fx and Fy Forces vs Time ({trial})')
    axs[row, col].legend()
    axs[row, col].grid(True)
    axs[row, col].set_ylim(-2, 4)  # Set y-axis limit

plt.tight_layout()
plt.savefig("image.png")
plt.show()

```
