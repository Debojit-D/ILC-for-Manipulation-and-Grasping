```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Define the range for x and y
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)

# Create meshgrid for x and y
X, Y = np.meshgrid(x, y)

# Define the function z = xy + 1
Z = X * Y + 1

# Create a 3D plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Plot the surface
ax.plot_surface(X, Y, Z, cmap='viridis')

# Labels and title
ax.set_xlabel('X axis')
ax.set_ylabel('Y axis')
ax.set_zlabel('Z axis')
ax.set_title('Surface Plot of z = xy + 1')

# Show the plot
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Define the range for x and y
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)

# Create meshgrid for x and y
X, Y = np.meshgrid(x, y)

# Define the function z = xy + 1
Z = X * Y + 1

# Define the integer value k for the contour, and calculate the contour level
k = 2  # Example integer value
contour_level = k - 1  # This translates to xy = k - 1

# Create a 3D plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Plot the surface
surface = ax.plot_surface(X, Y, Z, cmap='viridis', alpha=0.7, label='z = xy + 1')

# Plot the contour for xy + 1 = k
# We find all points where Z is close to k
contour_mask = np.abs(Z - k) < 0.1
ax.scatter(X[contour_mask], Y[contour_mask], Z[contour_mask], color='red', s=1)

# Labels and title
ax.set_xlabel('X axis')
ax.set_ylabel('Y axis')
ax.set_zlabel('Z axis')
ax.set_title('Surface Plot of z = xy + 1 with Contour at z = {}'.format(k))

# Show the plot
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from matplotlib.lines import Line2D  # Import to create custom legend entries

# Define the range for x and y
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)

# Create meshgrid for x and y
X, Y = np.meshgrid(x, y)

# Define the function z = xy + 1
Z = X * Y + 1

# Define the list of integer values for the contours
k_values = [3, 5, 7, 10]  # Example list of integer levels

# Create a 3D plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Plot the surface
surface = ax.plot_surface(X, Y, Z, cmap='viridis', alpha=0.7)

# Plot the contours for each value in k_values
colors = ['red', 'blue', 'green', 'black']  # Define a list of colors for the contours
legend_elements = []  # List to hold the legend entries

for k, color in zip(k_values, colors):
    contour_level = k - 1  # This translates to xy = k - 1
    contour_mask = np.abs(Z - k) < 0.1  # Find all points where Z is close to k
    scatter = ax.scatter(X[contour_mask], Y[contour_mask], Z[contour_mask], color=color, s=50)
    legend_elements.append(Line2D([0], [0], marker='o', color='w', label=f'z = {k}', markerfacecolor=color, markersize=10))

# Labels, title, and custom legend
ax.set_xlabel('X axis')
ax.set_ylabel('Y axis')
ax.set_zlabel('Z axis')
ax.set_title('Surface Plot of z = xy + 1 with Multiple Contours')
ax.legend(handles=legend_elements)

# Show the plot
plt.show()
```
